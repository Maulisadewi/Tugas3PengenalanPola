```python
# import StemmerFactory class
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming_list_of_words (l):
    """
    untuk stemming list yang berisi words
    arguments : l         -- list
    returns   : list_stem -- list dari kata yang telah distemming
    """
    words_stem = lambda x: stemmer.stem(x)
    list_stem = list(map(words_stem, l))
    return list_stem
```

```python
def sorted_words_in_string(String):
    """
    Arguments:
        String -- inputan berupa string (dapat memuat beberapa kata yang dipisahkan dengan spasi
    
    Output:
        kata-kata dalam string terurut
    """
    words = String.split()
    words.sort()
    sorted_sentence = ' '.join(word for word in words)
    return sorted_sentence
```

```python
def sorted_words_independent(l):
    """
    """
    new_list = list(map(sorted_words_in_string,l))
    return new_list
```
